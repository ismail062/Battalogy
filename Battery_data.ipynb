```python
import pandas as pd
from rdflib import Namespace, Graph, URIRef, RDF, RDFS, Literal
```

```python
filepath = "Student Results/A123/A123 drive cycle 20degC.csv"
ontology_file = "Battology_population.owl"
table_name = "Battery_Test"
df = pd.read_csv(filepath)
df.head()
```

```python
def extend_owl_turtle(owl_file, triples_to_add):
    # Create an RDF graph
    graph = Graph()

    # Load the existing OWL Turtle file
    print(owl_file, graph)
    try:
        graph.parse(owl_file, format="turtle")
    except FileNotFoundError:
        print(f"Error: {owl_file} not found.")
        return

    # Add new triples to the graph
    for triple in triples_to_add:
        subject, predicate, obj = triple
        print({subject: subject, predicate: predicate, obj: obj})
        graph.add((subject, predicate, obj))

    # Save the updated graph to the OWL Turtle file
    graph.serialize(destination=owl_file, format="turtle")
```

```python
extend_owl_turtle(ontology_file, [])
```

```python
ontologyTriples = []
ns = Namespace('http://www.semanticweb.org/v0/batteryOntology/')

hasAmpHour = URIRef(ns['hasTestAmpHour'])
hasCurrent = URIRef(ns['hasTestCurrent'])
hasMode = URIRef(ns['hasTestMode'])
hasPower = URIRef(ns['hasTestPower'])
hasStep = URIRef(ns['hasTestStep'])
hasStepTime = URIRef(ns['hasTestStepTime'])
hasTemp1 = URIRef(ns['hasTestTemprature1'])
hasTemp2 = URIRef(ns['hasTestTemprature2'])
hasTemp3 = URIRef(ns['hasTestTemprature2'])
hasTemp4 = URIRef(ns['hasTestTemprature4'])
hasVoltage = URIRef(ns['hasTestVoltage'])
hasWattHour = URIRef(ns['hasTestWattHour'])
batteryClass = URIRef(ns+'Battery_Test')

for index, row in df.iterrows():
    if(index > 0):
        test= URIRef(ns+'20degC_test'+str(index))
        ontologyTriples.append((test, RDF.type, batteryClass))
        
        step = Literal(int(row['Step']))
        ontologyTriples.append((test, hasStep, step))
        
        step_time = Literal(row['Step Time, S'])
        ontologyTriples.append((test, hasStepTime, step_time))

        current = Literal(row['Current, A'])
        ontologyTriples.append((test, hasCurrent, current))
        
        voltage = Literal(row['Voltage, V'])
        ontologyTriples.append((test, hasVoltage, voltage))
        
        power = Literal(row['Power, W'])
        ontologyTriples.append((test, hasPower, power))
        
        wattHour = Literal(row['Watt-Hours, WH'])
        ontologyTriples.append((test, hasWattHour, wattHour))
        
        ampHour = Literal(row['Amp-Hours, AH'])
        ontologyTriples.append((test, hasAmpHour, ampHour))
        
        temp1 = Literal(row['Temperature A1'])
        ontologyTriples.append((test, hasTemp1, temp1))
        
        temp2 = Literal(row['Temperature A2'])
        ontologyTriples.append((test, hasTemp2, temp2))
        
        temp3 = Literal(row['Temperature A3'])
        ontologyTriples.append((test, hasTemp3, temp3))
        
        temp4 = Literal(row['Temperature A4'])
        ontologyTriples.append((test, hasTemp4, temp4))
        
        mode = Literal(row['Mode'])
        ontologyTriples.append((test, hasMode, mode))
        # values = ", ".join([index, step, step_time, 1, '2024-01-06',current, voltage, power, ampHour, wattHour, temp1, temp2, temp3, temp4])
        # query = f"INSERT INTO {table_name} VALUES ({index}, {int(step)}, {int(step_time)}, 1, '2024-01-06', {current}, {voltage}, {power}, {wattHour}, {ampHour}, {temp1}, {temp2}, {temp3}, {temp4}, '{mode}', 'drive cycle 40degC.csv');"

        # ontologyTriples.append([test, step, step_time, current, voltage, power, wattHour, ampHour, temp1, temp2, temp3, temp4, mode])
```

```python
from rdflib import Graph, URIRef, Literal, Namespace, RDF, XSD
import csv


# df = pd.read_csv('corruption_letter.csv')
g = Graph()
g.parse('CaseControlStudy.rdf')
ns = Namespace('http://example.org/CaseControlStudy#')
rdf = Namespace('http://www.w3.org/1999/02/22-rdf-syntax-ns#')
rdfs = Namespace('http://www.w3.org/2000/01/rdf-schema#')

with open('corruption_letter.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file)
    count = 1
    next(csv_reader)
    for  row in csv_reader:
        new_country = URIRef(ns+(row[2].replace(' ', '_')))
        new_corruption = URIRef(ns+'Corruption_Letter_'+str(count))
        print(new_country, new_corruption)
        g.add((new_country, RDF.type, ns.Country))
        g.add((new_corruption, RDF.type, ns.Corruption_Letter))
        # subject = URIRef(ns[row[0]])
        cpi_data = URIRef(ns['contain_CPI_Data'])
        cp_vacinated = URIRef(ns['contain_Percentage_Vaccinated'])
        hdi_data = URIRef(ns['contain_HDI_Data'])

        #belongs to
        belong_to = URIRef(ns['belong_To'])

        g.add((new_corruption, belong_to, new_country))

        cpi_obj = Literal(row[0], datatype=XSD.float)
        vacination_obj = Literal(row[1], datatype=XSD.float)
        hdi_obj = Literal(row[3], datatype=XSD.float)

        g.add((new_corruption, cpi_data, cpi_obj))
        g.add((new_corruption, cp_vacinated, vacination_obj))
        g.add((new_corruption, hdi_data, hdi_obj))
        count += 1

g.serialize(destination='CaseControlStudy.rdf', format='xml')
```
